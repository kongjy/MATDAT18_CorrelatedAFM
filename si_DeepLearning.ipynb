```python
## Function to create consistent figures ##

def show(img, cmap = 'plasma', colorbar = True):
    """Displays image with colorbar."""
    fig = plt.imshow(img, cmap = cmap)
    imgstd = img.std()
    imgmean = img.mean()
    ll = imgmean - (2*imgstd)
    ul = imgmean + (2*imgstd)
    plt.clim(ll,ul)
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    if colorbar == True:
        plt.colorbar()
    return 
```

## Prepare the data

Begin by stratifying the data into smaller subsets.

```python
## Scale the response ##
Ys = scale(cAFM_reg)

## Stratify the data ##
strat_dim = 7 #19
strat_n = [int(i/strat_dim) for i in cAFM_reg.shape]
print(strat_n)
X = np.zeros((np.prod(strat_n),strat_dim,strat_dim,hyper.shape[2]))
Y = np.zeros((np.prod(strat_n),strat_dim,strat_dim))

for i in range(strat_n[0]):
    for j in range(strat_n[1]):
        X[j+i*strat_n[1],:,:,:] = hyper[(i*strat_dim):((i+1)*strat_dim), (j*strat_dim):((j+1)*strat_dim),:]
        Y[j+i*strat_n[1],:,:] = Ys[(i*strat_dim):((i+1)*strat_dim), (j*strat_dim):((j+1)*strat_dim)]
        


M = X.shape[0]
print(X.shape)

Y = np.expand_dims(Y,axis=3)
print(Y.shape)
```

```python
#------------------------------------------------------#
#--- Determine sample sizes and shuffle the indeces ---#

test_n = test_n = np.sort(np.add.outer(np.arange(strat_n[1]-7,strat_n[1]), np.arange(0, M, strat_n[1])).ravel()).tolist()
val_n = list(set(np.arange(X.shape[0])) - set(test_n))


# percentages used = 0.6, 0.15, 0.25
div = np.array([.8, .20])

n = len(val_n)*div
n = n.astype(int)
n[0] = n[0] + len(val_n) - sum(n)
print(n, len(test_n))

```

```python
#-----------------------------------------------------------#
#--- Divide the data into training, test, and validation ---#

X_train, X_val, X_test = X[val_n[:n[0]],:,:,:], X[val_n[n[0]:sum(n)],:,:,:], X[test_n,:,:,:]
Y_train, Y_val, Y_test = Y[val_n[:n[0]],:,:,:], Y[val_n[n[0]:sum(n)],:,:,:], Y[test_n,:,:,:]


print(X_train.shape)
print(X_val.shape)
```

## Define the CNN Model

```python
#---------------------------#
#-------- CNN Model --------#

X_input = Input(shape=X_train.shape[1:])
x = UpSampling2D((4, 4))(X_input)


## Encoder ##
# Block 1
x = Conv2D(16, (3,3), padding='same')(x)
#x = BatchNormalization()(x)
x = Activation('relu')(x)
if False:
    x = Conv2D(16, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPool2D((2, 2))(x)


if True:
    # Block 2
    if False:
        x = Conv2D(32, (3,3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
    x = Conv2D(32, (3,3), padding='same')(x)
#    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPool2D((2, 2))(x)
    


    # Block 3
    if False:
        x = Conv2D(64, (3,3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
    x = Conv2D(64, (3,3), padding='same')(x)
#    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPool2D((2, 2))(x)




## Decoder ##
# Block 4
if True:
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(64, (3,3), padding='same')(x)
#    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    if False:
        x = Conv2D(64, (3,3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)


    # Block 5
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(32, (3,3), padding='same')(x)
#    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    if False:
        x = Conv2D(32, (3,3), padding='same')(x)
#        x = BatchNormalization()(x)
        x = Activation('relu')(x)


    # Block 6
    if True:
#        x = UpSampling2D((2, 2))(x)
        x = Conv2D(16, (3,3), padding='same')(x)
#        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        if False:
            x = Conv2D(16, (3,3), padding='same')(x)
            x = BatchNormalization()(x)
            x = Activation('relu')(x)




# Block 5
x = MaxPool2D((4, 4))(x)
x = Conv2D(Y_test.shape[3], (3,3), padding='same')(x)
#x = BatchNormalization()(x)
x = Activation('linear')(x)


model = Model(inputs=[X_input], outputs=[x])
model.summary()
```

```python
## CNN Inputs ##
lr = 0.0001
loss = 'mean_squared_error'
```

```python
#------------------------------------#
#------- Analysis and Results -------#

lr_str = re.sub("\\.","", '%.0E' %lr)
opt = RMSprop(lr=lr)

chkpoint = ModelCheckpoint('out_' + str(np.prod(strat_n)) + 'samples_' + lr_str + '_' + loss + '.h5', save_best_only=True)

model.compile(optimizer=opt, loss=loss, metrics=['mae'])


## Train NN ##
t0 = time()
model.fit(X_train, Y_train, epochs=10, validation_data=(X_val, Y_val), callbacks=[chkpoint], batch_size=37)

print("Done in %0.3fs." % (time() - t0))
```

```python
## Obtain predictions ##

t0 = time()
preds = model.predict(X_test)
print("Done in %0.3fs." % (time() - t0))
print(preds.shape)
```

```python
## Combine into single predicted image ##

split_preds = np.zeros((strat_n[0], int(preds.shape[0]/strat_n[0]), strat_dim, strat_dim, preds.shape[3]))

for i,m in enumerate(test_n):
    split_preds[np.floor(m/strat_n[1]).astype(int), m%strat_n[1]-(strat_n[1]-split_preds.shape[1]), :, :, :] = preds[i,:,:,:]

print(split_preds.shape)
    

preds_image = np.zeros((split_preds.shape[0]*split_preds.shape[2], split_preds.shape[1]*split_preds.shape[3], split_preds.shape[4]))
for i in range(split_preds.shape[0]):
    for j in range(split_preds.shape[1]):
        preds_image[(i*strat_dim):((i+1)*strat_dim),(j*strat_dim):((j+1)*strat_dim),:] = split_preds[i,j,:,:,:]
        

print(preds_image.shape)
```

```python
tlocs = (cAFM_reg.shape[1] - preds_image.shape[1])

plt.figure(figsize=(12,10))
plt.subplot(131), show(Ys[:,tlocs:]), plt.title("Truth")
plt.subplot(132), show(preds_image[:,:,0]), plt.title("CNN Predictions")
plt.subplot(133), show(preds_image[:,:,0] - Ys[:,tlocs:]), plt.title("Error")
plt.show()
```
